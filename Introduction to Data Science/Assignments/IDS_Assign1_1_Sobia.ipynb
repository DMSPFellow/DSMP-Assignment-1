```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
r = requests.get(url)
r.content
```

```python
soup = BeautifulSoup(r.text, "lxml")
soup
```

```python
soup.prettify()
```

```python
soup
```

```python
names = soup.find_all("a", class_="title")
print(names)
```

```python
Names = []
for i in names:
    print(i.text)
    Names.append(i.text)
```

```python
prices = soup.find_all("h4", class_="pull-right price")
print(prices)
```

```python
Prices = []
for i in prices:
    print(i.text)
    Prices.append(i.text)
```

```python
description = soup.find_all("p", class_="description")
print(description)
```

```python
Description = []
for i in description:
    print(i.text)
    Description.append(i.text)
```

```python
reviews = soup.find_all("p", class_="pull-right")
print(reviews)
```

```python
Reviews = []
for i in reviews:
    print(i.text)
    Reviews.append(i.text)
```

```python
print(Names)
print(len(Names))

print(Prices)
print(len(Prices))

print(Description)
print(len(Description))

print(Reviews)
print(len(Reviews))
```

```python
dataset = {
    "Name": Names,
    "Price": Prices,
    "Description": Description,
    "Reviews": Reviews
}
```

```python
df = pd.DataFrame(dataset)
df
```
